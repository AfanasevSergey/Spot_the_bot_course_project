```python
import requests, bs4, os, errno, zipfile, glob
from urllib.request import urlretrieve
import pandas as pd
import nltk
import codecs
```

```python
if not os.path.exists('books_it/'):
    try:
        os.makedirs('books_it/')
    except OSError as e:
        if e.errno != errno.EEXIST:
            raise
```

```python
urls_to_books = []

if not os.path.exists('urls_to_books_it.txt'):
    page_w_books_url = 'http://www.gutenberg.org/robot/harvest?filetypes%5B%5D=txt&langs%5B%5D=it&#39;'
    
    count_page = 0
    while 1 == 1:
        is_last_page = False
        print('Reading page: ' + page_w_books_url)
        page_w_books = requests.get(page_w_books_url, timeout=20.0)
        
        if page_w_books:
            page_w_books = bs4.BeautifulSoup(page_w_books.text, "lxml")
            urls = [el.get('href') for el in page_w_books.select('body > p > a')]
            url_to_next_page = page_w_books.find_all('a', string='Next Page')
            
            if len(urls) > 0:
                urls_to_books.append(urls)

                if url_to_next_page[0]:
                    page_w_books_url = "http://www.gutenberg.org/robot/" + url_to_next_page[0].get('href')
            else:
                is_last_page = True

        count_page =+ 1
        
        if is_last_page or count_page > 500:
            break
```

```python
urls_to_books = [item for sublist in urls_to_books for item in sublist]
# Backing up the list of URLs

with open('urls_to_books_it.txt', 'w') as output:
    for u in urls_to_books:
        output.write('%s\n' % u)
```

```python
print("Books count: ", len(urls_to_books))
```

```python
if os.path.exists('urls_to_books_it.txt') and len(urls_to_books) == 0:
    with open('urls_to_books_it.txt', 'r') as f:
        urls_to_books = f.read().splitlines()
```

```python
#collect books
url_num = 0
book_dict = dict()

for url in urls_to_books[url_num:]:
    dst = 'books_it/' + url.split('/')[-1].split('.')[0].split('-')[0]
    
    if url.find('http://') > -1:
    
        with open('logfile.log', 'w') as f:
            f.write('Unzipping file #' + str(url_num) + ' ' + dst + '.zip' + '\n')

        if len(glob.glob(dst + '*')) == 0:
            urlretrieve(url, dst + '.zip')

            with zipfile.ZipFile(dst + '.zip', "r") as zip_ref:
                try:
                    zip_ref.extractall("books_it/")
                    print(str(url_num) + ' ' + dst + '.zip ' + 'unzipped successfully!')
                    book_dict[url_num] = dst + '.zip'
                except NotImplementedError:
                    print(str(url_num) + ' Cannot unzip file:', dst)

            os.remove(dst + '.zip')

        url_num += 1
```
